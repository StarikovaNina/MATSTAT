```python
import numpy as np
import random as rd
from math import sqrt

# парамеры
theta = 1          # значение параметра
N = 100            # объем выборки
beta = 0.95        # доверительная вероятность
B = 1000           # число бутстрап-повторений

def p(x: float) -> float:
    return 1 / theta if theta <= x <= 2 * theta else 0

def F(x: float) -> float:
    if x < theta:
        return 0
    elif theta <= x <= 2 * theta:
        return x / theta - 1
    else:
        return 1

def F_inv(y: float) -> float:
    return theta * (1 + y)


# генерация выборки
arr = []
for _ in range(N):
    y = rd.random()
    arr.append(F_inv(y))

arr = np.array(arr)


# оценки
x_mean = np.mean(arr)
x_max = np.max(arr)

# ОММ
theta_mm = 2 * x_mean / 3

# ОМП
theta_mle = x_max / 2

# исправленная ОМП
theta_mle_corr = (N + 1) * x_max / (2 * N + 1)

print(f"theta = {theta}")
print(f"Среднее выборки = {x_mean}")
print(f"Максимум выборки = {x_max}")
print()

print(f"Оценка ОММ = {theta_mm}")
print(f"Оценка ОМП = {theta_mle}")
print(f"Исправленная ОМП = {theta_mle_corr}")
print()

# точный доверительный интервал
lower_exact = x_max / (1 + ((1 + beta) / 2) ** (1 / N))
upper_exact = x_max / (1 + ((1 - beta) / 2) ** (1 / N))

print("Точный доверительный интервал:")
print(f"({lower_exact}, {upper_exact})")
print(f"Длина: {upper_exact - lower_exact}")
print()


# асимптотический доверительный интервал
z = 1.96

lower_asymp = theta_mm - z * theta_mm / (3 * sqrt(3 * N))
upper_asymp = theta_mm + z * theta_mm / (3 * sqrt(3 * N))

print("Асимптотический доверительный интервал:")
print(f"({lower_asymp}, {upper_asymp})")
print(f"Длина: {upper_asymp - lower_asymp}")
print()

# бутстраповский доверительный интервал
h_wave = (N + 1) * x_max / (2 * N + 1)

bootstrap_delta = []

for _ in range(B):
    sample = np.random.choice(arr, size=N, replace=True)
    bootstrap_delta.append((N + 1) * np.max(sample) / (2 * N + 1) - h_wave)

variation_row = sorted(bootstrap_delta)

t_1 = variation_row[int((1 - beta) * B / 2)]
t_2 = variation_row[int((1 + beta) * B / 2)]


lower_bound = h_wave - t_2
upper_bound = h_wave - t_1

print("Бутстраповский доверительный интервал:")
print(f"({lower_bound}, {upper_bound})")
print(f"Длина: {upper_bound - lower_bound}")
print()


# сравнение
intervals = {
    "Точный": (lower_exact, upper_exact, upper_exact - lower_exact),
    "Асимптотический": (lower_asymp, upper_asymp, upper_asymp - lower_asymp),
    "Бутстраповский": (lower_bound, upper_bound, upper_bound - lower_bound),
}

print("Сравнение интервалов:")


best = sorted(intervals.items(), key=lambda x: x[1][2])

print(f"{best[0][0]} < {best[1][0]} < {best[2][0]}")
print()
print(f"Самый короткий интервал: {best[0][0]}")
print()
print("Вывод, самым коротким в зависимости от выборки будет бутсраповский или точный, ")
print("потому что метод бутстрапа занижает длину доверительного интервала, ")
print("а точный метод даёт лучшую оценку, чем асимптотический")
```
